```python
# Mengimport Library
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Integral
def func(x):
    return (x**-3) + np.cos(x)
a = 1.0
b = 5.0
```

```python
# Metode Trapezoid
n = 10000
dx = (b-a)/(n-1)
x = np.linspace(a,b,n)

sigma = 0
for i in range (1, n-1):
    sigma += func(x[i])

hasil = 0.5*dx*(func(x[0])+2*sigma+func(x[-1]))

print(hasil)
```

```python
xp =np.linspace(a,b,10000)
plt.plot(xp,func(xp))
plt.show()
```

```python
xp =np.linspace(a,b,10000)
plt.plot(xp,func(xp))

for i in range (n):
    plt.bar(x[i],func(x[i]),align = 'edge',width = 0.000001, edgecolor='black')

plt.show()
```

```python
# Mengimport Library
import numpy as np
import matplotlib.pyplot as plt

# Fungsi yang akan diintegralkan
def func(x):
    return (x**-3) + np.cos(x)

# Batas Integrasi
a = 1.0   # Batas Bawah
b = 5.0  # Batas Atas
n = 10000  # Jumlah grid, harus ganjil untuk metode Simpson
```

```python
# Simpson's Rule
if n % 2 == 0:
    n += 1

x = np.linspace(a, b, n)
dx = (x[-1] - x[0]) / (n - 1)

hasil = func(x[0]) + func(x[-1])

for i in range(1, n-1, 2):
    hasil += 4 * func(x[i])

for i in range(2, n-2, 2):
    hasil += 2 * func(x[i])

hasil *= dx / 3

#Visualisasi grafik dan bar
xp = np.linspace(a,b,10000)
plt.plot(xp, func(xp))

for i in range(n):
    plt.bar(x[i], func(x[i]), align='edge', width=dx, color='red', edgecolor='black')
plt.show()
print(hasil)
```
